```python
columns = ["epic_candname",
           "pl_name",
           "k2c_refdisp",
           "k2c_disp",
           "pl_orbper", #Orbital Period [days]
           "pl_tranmid", #The time given by the average of the time the candidate begins to cross the stellar limb and the time the candidate finishes crossing the stellar limb.
           "pl_trandur", # 	Transit Duration [days]
           "pl_imppar", #Impact Parameter
           "pl_orbincl", #Inclination [deg]
            "pl_ratdor", #Ratio of Distance to Stellar Radius, a/R
           "pl_ratror", #Ratio of Candidate to Stellar Radius, r/R
           "pl_eqt", # Equilibrium Temperature [K] 
           "pl_trandep", #
           "pl_rade", #Candidate Radius [Earth radii] 
           
           "st_teff", #Effective Temperature [K]
           "st_logg", #Stellar Surface Gravity
           "st_metfe", #Metallicity Ratio
           "st_rad" # Stellar Radius [Solar radii]
          ]

#columnas que no estan: "Fitted Stellar Density", "Semi-major Axis"
#"Limb Darkening Coeff1", se puede sacar de cross matchin con catalogos de las estrellas
              
df_meta= df_aux[columns]
df_meta.columns = ["EPIC Name","K2 Name", "K2 Disposition", "NExScI Disposition", 
                   "Period", "First Transit Time (BJD)", "Duration", "Impact Parameter", "Inclination",
                   "a/R", "r/R", "Teq",
                  "Transit Depth",  "Planet Radius", 
                   'Teff', 'log(g)', 'Metallicity', 'Stellar Radius'
                  ]
df_meta["Time of Transit Epoch"] = df_meta["First Transit Time (BJD)"] - 2454833 #K2 reference..

df_meta.to_csv(folder+"K2_CAND_metadata.csv",index=False)
df_meta.head()
```

```python
#borrar nans... arreglo variable
coupled_lc = []
coupled_time = []
delta_time = []
for i in range(N):
    mask_nan_aux = np.isnan(process_lc[i])
    coupled_lc.append(process_lc[i][~mask_nan_aux])
    
    time = lc_time_aux[i][~mask_nan_aux]
    # calculate delta time --> this could be done after padding is done..
    delta_time.append(np.hstack([[0],np.diff(time)]))
    coupled_time.append(time)
    
coupled_lc = np.asarray(coupled_lc)
coupled_time = np.asarray(coupled_time)
delta_time = np.asarray(delta_time)
coupled_time
```

```python
T = 300 # CANTIDAD DE PUNTOS EN LA CURVA GLOBAL

X_fold_lc = []
X_fold_lc_aug = [] #con curvas invertidas
X_fold_time = []
for n in range(N):
    if n%500 == 0:
        print("Va en el, ",n)
            
    per, t0, dur = df_meta[["Period", "Time of Transit Epoch","Duration"]].values[n]
    t = coupled_time[n]
    y = coupled_lc[n]

    t_fold, val_fold = phase_fold_and_sort_light_curve(t, y, period= per, t0= t0)

    val_glo = global_view(t_fold, val_fold, period=per, num_bins=T, bin_width_factor=1 / T)
    X_fold_lc.append(val_glo)

    #dos opciones para revertir
    t_fold_aug = t_fold[::-1]*-1 #importante!
    val_fold_aug = val_fold[::-1]

    val_glo = global_view(t_fold_aug, val_fold_aug, period=per, num_bins=T, bin_width_factor=1 / T)
    X_fold_lc_aug.append(val_glo)

    t_glo_diff = np.concatenate([[0], np.tile(per/T, T-1)])  #np.diff(t_glo)
    X_fold_time.append(t_glo_diff) #differences values

X_fold_lc = np.asarray(X_fold_lc)
X_fold_lc_aug = np.asarray(X_fold_lc_aug)
X_fold_time = np.asarray(X_fold_time)
X_fold_lc.shape
```
